## Dachshund-Badger Simulation
A well know fact is that dachshunds were bred to be world-class badger hunters.  In this exercise, you will create a simple simulation of a dachshund-badger population over time. To model the dachshund/badger population, you will simulate a simple predator-prey interaction using the Lotka-Volterra equations (a common model in biology). 

**Badger growth rate:**
Badger(t+1) =Badger(t)+(α⋅Badger(t))−(β⋅Badger(t)⋅Dachshund(t))

**Dachshund growth rate:**
Dachshund(t+1)=Dachshund(t)+(δ⋅Badger(t)⋅Dachshund(t))−(γ⋅Dachshund(t))

Where:
 * α: Badger birth rate

 * β: Death rate of badgers (depends on the predator)

 * δ: Dachshund reproduction rate (based on consumed badgers)

 * γ: Dachshunds death rate

(Note: In reality, dachshunds typically just corner badgers rather than eating them, but they're known for their ferocity, so I wouldn’t put it past them.
```python
# 10pts Function that calculates one step of the dachshund/badger equation.

# returns the current badger and dachshund populations after applying the equations
def current_badger_and_dachshund_count(prev_badger_num, prev_dachshund_num, badger_birth_rate, badger_death_rate, dachshund_birth_rate, dachshund_death_rate):
    # calculates the number of badgers
    badger_count = prev_badger_num + (badger_birth_rate * prev_badger_num) - (badger_death_rate * prev_badger_num * prev_dachshund_num)

    # handles fraction and negative numbers of badgers
    badger_count = round(badger_count)
    if (badger_count < 0):
        badger_count = 0
    
    # calculates the number of dachshunds
    dachshund_count = prev_dachshund_num + (dachshund_birth_rate * prev_badger_num * prev_dachshund_num) - (dachshund_death_rate * prev_dachshund_num)

    # handles fraction and negative numbers of dachshunds
    dachshund_count = round(dachshund_count)
    if (dachshund_count < 0):
        dachshund_count = 0

    return badger_count, dachshund_count
```

```python
#15pts How long can badgers survive?

# hard coded initial values for current_badger_and_dachshund_count parameters
prev_badger_num = 200
prev_dachshund_num = 1
badger_birth_rate = 0.2
badger_death_rate = 0.005
dachshund_birth_rate = 0.001
dachshund_death_rate = 0.2

# returns the current badger and dachshund populations after applying the equations
def current_badger_and_dachshund_count(prev_badger_num, prev_dachshund_num, badger_birth_rate, badger_death_rate, dachshund_birth_rate, dachshund_death_rate):
    # calculates the number of badgers
    badger_count = prev_badger_num + (badger_birth_rate * prev_badger_num) - (badger_death_rate * prev_badger_num * prev_dachshund_num)

    # handles fraction and negative numbers of badgers
    badger_count = round(badger_count)
    if (badger_count < 0):
        badger_count = 0
    
    # calculates the number of dachshunds
    dachshund_count = prev_dachshund_num + (dachshund_birth_rate * prev_badger_num * prev_dachshund_num) - (dachshund_death_rate * prev_dachshund_num)

    # handles fraction and negative numbers of dachshunds
    dachshund_count = round(dachshund_count)
    if (dachshund_count < 0):
        dachshund_count = 0

    return badger_count, dachshund_count

# prints out the years and their respective badger and dachshund populations
def print_populations():
    # explicitly tell python to use the globals
    global prev_badger_num, prev_dachshund_num
    year = 1

    # prints out the initial number of badgers and dachshunds
    print(f"t = 0:  {prev_badger_num} badgers,  {prev_dachshund_num} dachshunds")

    # prints out the number of badgers and dachshunds each year for a given number of years
    while prev_badger_num != 0:
        #collects calculated population numbers of badgers and dachshunds
        badger_count, dachshund_count = current_badger_and_dachshund_count(prev_badger_num, prev_dachshund_num, badger_birth_rate, badger_death_rate, dachshund_birth_rate, dachshund_death_rate)

        # prints out the year along with the population counts
        print(f"t = {year}: {badger_count} badgers, {dachshund_count} dachshunds")

        # updates the populations of badgers and dachshunds for the next iteration of the loop
        prev_badger_num = badger_count
        prev_dachshund_num = dachshund_count
        year += 1
    
    return year

def main():
    global prev_badger_num, prev_dachshund_num
    largest_years = 0
    for i in range(1, 6):
        years = print_populations()
        prev_badger_num = 200
        prev_dachshund_num = i
        if years > largest_years:
            largest_years = years
    
    print(f"With {prev_dachshund_num} dachshunds at time 0, badgers will survive for {years} years, which is the maximum possible years that 200 badgers could survive.")

main()
```
